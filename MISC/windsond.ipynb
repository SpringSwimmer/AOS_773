```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
wsd_1827 = pd.read_csv('/home/leoseen/aos_773/windsond_data/2026-03-03_1827.csv')
wsd_1948 = pd.read_csv('/home/leoseen/aos_773/windsond_data/2026-03-03_1948_1.csv')
```

```python
print(wsd_1827.keys())
```

```python
fig, ax = plt.subplots(1,1,figsize = (10,8))
ax.plot(wsd_1827[' Temperature (C)'], wsd_1827['Height (m AGL)'], color = 'black', linewidth = 3)
ax.set_xlim(0,5)
ax.set_ylim(0,1000)
ax.tick_params(axis = 'both', labelsize = 16)
ax.set_xlabel(r'Temperature [C$^o$]', fontsize = 20)
ax.set_ylabel(r'Height [m]', fontsize = 20)
ax.grid(alpha = 0.4, color = 'grey', linestyle = '--')
ax.grid(alpha = 0.4, color = 'grey', linestyle = '--')
# ax.legend(fontsize = 20)
```

```python
fig, ax = plt.subplots(1,1,figsize = (10,8))
ax.plot(wsd_1827[' Wind speed (m/s)'], wsd_1827['Height (m AGL)'], color = 'black', linewidth = 3)
ax.plot(wsd_1948[' Wind speed (m/s)'], wsd_1948['Height (m AGL)'], color = 'black', linewidth = 3)
ax.set_xlim(0,5)
ax.set_ylim(0,1000)
ax.tick_params(axis = 'both', labelsize = 16)
ax.set_xlabel(r'Wind Speed [m/s]', fontsize = 20)
ax.set_ylabel(r'Height [m]', fontsize = 20)
ax.grid(alpha = 0.4, color = 'grey', linestyle = '--')
ax.grid(alpha = 0.4, color = 'grey', linestyle = '--')
# ax.legend(fontsize = 20)
```
