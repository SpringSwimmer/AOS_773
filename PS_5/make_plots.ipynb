```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
```

```python
time = np.arange(0,43200+3600, 3600)
nocturnal_winds_u = 10+3*np.sin(1e-4*time)-3*np.cos(1e-4*time)
nocturnal_winds_v = 3*np.sin(1e-4*time)+3*np.cos(1e-4*time)
```

```python
print(np.max(nocturnal_winds_u))
print(np.where(nocturnal_winds_u == np.max(nocturnal_winds_u))[0][0])
```

```python
v = np.sqrt(nocturnal_winds_u**2+nocturnal_winds_v**2)
print(np.max(v))
print(np.where(v == np.max(v))[0][0])
```

```python
print(np.where(v>10)[0])
```

```python
fig, ax = plt.subplots(1,1,figsize = (10,8))
ax.plot(time/3600, nocturnal_winds_u, linewidth = 3, color = 'red')
ax.set_xlim(0,12)
ax.set_title(r'Fu$_\mathrm{day} = $Fv$_\mathrm{day} = 3$ m/s,  $f = 10^{-4} \mathrm{ s}^{-1}$, $U_g = 10$ m/s, $V_g = 0$ m/s', fontsize = 20)
ax.set_xlabel(f'Time since sunset [hr]', fontsize = 20)
ax.set_ylabel(f'Nocturnal Wind U [m/s]', fontsize = 20)
ax.tick_params(axis = 'both', which = 'major', labelsize = 16, width = 1, length = 10)
ax.tick_params(axis = 'both', which = 'minor', labelsize = 16, width = 1, length = 4)
ax.xaxis.set_minor_locator(MultipleLocator(1))
ax.grid(alpha = 0.4, color = 'grey', linestyle = '--')
```

```python
fig, ax = plt.subplots(1,1,figsize = (10,8))
ax.scatter(nocturnal_winds_u, nocturnal_winds_v, linewidth = 3, color = 'red')
# ax.set_xlim(0,12)
ax.set_title(r'Fu$_\mathrm{day} = $Fv$_\mathrm{day} = 3$ m/s,  $f = 10^{-4} \mathrm{ s}^{-1}$, $U_g = 10$ m/s, $V_g = 0$ m/s', fontsize = 20)
ax.set_xlabel(f'Nocturnal Wind U [m/s]', fontsize = 20)
ax.set_ylabel(f'Nocturnal Wind V [m/s]', fontsize = 20)
ax.tick_params(axis = 'both', which = 'major', labelsize = 16, width = 1, length = 10)
ax.tick_params(axis = 'both', which = 'minor', labelsize = 16, width = 1, length = 4)
ax.xaxis.set_minor_locator(MultipleLocator(1))
for i in range(len(nocturnal_winds_u)):
    ax.annotate(i, [nocturnal_winds_u[i], nocturnal_winds_v[i]+0.2], fontsize = 20)
ax.grid(alpha = 0.4, color = 'grey', linestyle = '--')
ax.set_ylim(-5,5)
```
